Section 6.1 showed that the flux -- the flow across an infinitesimal surface -- is $$ d\Phi_v = {\bf v} \cdot d{\bf S}.$$  

Section 6.2 says that the same flux is also $$ d\Phi_v = \left(\nabla\cdot {\bf v}\right)dV.$$ 

Chapter 8 finally combines these two to state the Theorem of Gauss in integral form:
$$ \oint_S {\bf v} \cdot d{\bf S} =\int_V \left(\nabla\cdot {\bf v}\right)dV.$$

### Gauss' Theorem across different dimensions
The image above implies a 3D volume, but we have already seen -- and used -- lower dimensional versions of this.
The 2D version in Chapter 6 returns in Chapter 9: the Theorem of Stokes. There, $dV$ will be an area, and the surface reduces to a line integral. 

In 1D, the $dV$ is reduced to a line, and the $dS$ is formed by its two end points. 
Let's align the line with the x-axis, and then the 1D Gauss' Theorem is the familiar definition of integration:
$$ v_x(b) - v_x(a) = \int_a^b \frac{\partial v_x}{\partial x} dx.$$

### Gravity inside and outside a sphere
Let us consider the vector field that is the graviational field from a sphere with a radial density profile, so that ${\bf g(\bf{r})} = g(r)\hat{{\bf r}}$, where the unit vector points outward from the centre.

In Section 8.2, you will have worked your way through the problem of the gravitational field outside such a mass:
$${\bf g}(r) =- \frac{GM}{r^2}\hat{\bf r},$$

but Gauss' Theorem lets us express the gravitational field *inside* a sphere of radius $R$. If the distribution of density is radially symmetric, analytic solutions are straightforward. For example, the gravitational field inside a sphere with a *constant density* is
$${\bf g}(r) = -\frac{GMr}{R^3}\hat{\bf r}.$$

Combining these solutions, we know the gravitational acceleration by a homogeneous sphere from everywhere inside the sphere ($r<R$), to outside ($r>R$). In the following code we take values that represent the average radius of the Earth and its estimated mass:

```python
import numpy as np
import matplotlib.pyplot as plt

# The constants in SI units:
G = 6.67*10**(-11) # The Gravitational constant in m^3 s^(-2) kg^-1
M = 6*10**24 # The Earth's mass in kg
R = 6371*10**3 # The radius the Earth in m

#  A function to compute the magnitude of the (radial) gravitational acceleration, inside or outside the Earth:
def gravity(r):
    if r>R:
        g = M*G/r**2
    else:
        g =  M*G*r/R**3
    return g

# setting up an array with points from the centre (0) to well beyond the radius of the Earth:
rs= np.arange(0,10**7,1000)

# compute the magntiude of the gravitational acceleration for all points in the array rs:
gs = []
for r in rs:
    gs.append(gravity(r))

# plotting the results:
plt.plot(gs,rs/1000)
plt.axhline(R/1000,linestyle='dashed',color='r')
plt.xlabel("Gravitational acceleration (m/s$^2$)")
plt.ylabel("Distance to the centre of the Earth (km)")
plt.show()
```

We see that gravity at the centre of the Earth is zero, as any body there would be pulled in every direction equally, canceling all contributions. As we move to the surface, the gravitational field increases linearly in strength for this mass with constant density, until we reach the Earth's surface. The observed gravitational accelaration at the Earth's surface does not vary much from this simple model: around 9.8 m/s$^2$, or [980 Gals](https://en.wikipedia.org/wiki/Gal_(unit)). Slightly more accurate would be an ellipsoidal shape to the Earth, where gravity at the poles is about 9.83 and 9.78 m/s$^2$ at the equator.
