A scale analysis in Chapter 11 of [A Guided Tour of Mathematical Methods for the Physical Sciences](http://www.cambridge.org/nz/academic/subjects/physics/mathematical-methods/guided-tour-mathematical-methods-physical-sciences-3rd-edition#W7jZxE54KWWuUWEC.97) lets us treat the behaviour of waves in terms of ray theory. We found that this is not a bad approximation when the spatial variability in the physical parameters that define the wave propagation is large compared to the propagating wavelengths. Then, 
the Eikonal equation describes the travel time $T$ for a ray in a medium with wave speed $v({\bf r})$ as $$ |\nabla T({\bf r})|^2 = 1/v({\bf r})^2.$$ 
This describes how the gradient of the travel time $T$ of a ray depends on the (local) velocity, and in our book you can read how to derive [Snell's Law](https://en.wikipedia.org/wiki/Snell%27s_law) from the Eikonal equation. (Others first derive [Fermat's principle](
https://en.wikipedia.org/wiki/Fermat%27s_principle), and use it to prove Snell's Law) 

### Rays in a spherically symmetric Earth
Having accepted the representation of seismic rays in the Earth, we can estimate the travel time $T$ and epicentral distance $\Delta$ of these rays. To first order, the Earth is spherically symmetric, greatly simplifying ray propagation.In a spherically symmetric velocity model $v({\bf r})= v(r)$: 

![Figure 1](https://auckland.figshare.com/ndownloader/files/14738099)

First, a small part of the ray path $ds$ satisfies Pythagoras' Rule: $ds^2 = dr^2 + (rd\Delta)^2$. Secondly, The take-off angle $\sin i = dr/ds$. Thirdly, Snell's Law in a spherically symmetric Earth is $p = r\sin (i) /v$. This $p$ is the (constant) ray parameter. 

#### Homework
Combining these three, convince yourself that you can write:

$$ \frac{ds}{dr} = \frac{r}{\sqrt{r^2-(pv)^2}} \ \ \ \ \ \mbox{and}   \ \ \ \ \ \ \frac{d\Delta}{dr} = \frac{pv/r}{\sqrt{r^2-(pv)^2}}.$$
To find the total ray path length $s$ and the epicentral distance $\Delta$, we integrate these expressions with regards to $r$. 

Let's consider a ray that starts and ends at the surface:
![Figure 2](https://auckland.figshare.com/ndownloader/files/14754827)

Then, we take advantage of the symmetry, by integrating twice from the point of maximum depth of the ray $r=r_0$ to the surface $r= R$:
$$
\Delta(p) = 2 \int_{r_0}^R \frac{pv/r} {\sqrt{r^2-(pv)^2}}dr.
$$ 

The travel time requires integration of $\frac{1}{v}\frac{ds}{dr}$ over $r$, so that:
$$
T(p) = 2 \int_{r_0}^R \frac{r/v} {\sqrt{r^2-(pv)^2}}dr.
$$
These equations could have also been derived directly from the Eikonal equation and the equation of kinematic ray tracing (see Aki and Richards, for example).

### Rays in a sphere with constant speed
In his book *Principles of Seismology*, Udias derives directly for the special case of a sphere with a constant speed: $v(r) =c$:

![Figure 3](https://auckland.figshare.com/ndownloader/files/14754908)
In this simplest of cases, the rays don't bend and the results are
$$
T = \frac{2}{c}\sqrt{R^2-p^2c^2}
$$
and
$$
\Delta = 2\cos^{-1}\left(\frac{pc}{R}\right).
$$

####  Homework
Convince yourself of Udias' results by differentiating the expressions for $T$ and $\Delta$ with respect to $r$ to confirm the integrands for the general expressions of $T(p)$ and $\Delta(p)$ stated earlier.

Alternatively, we can compare the analytic solution and the intregral solution with a numerical approximation:

```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt

# integrand for the travel time:
def integrand(r, c, p):
    return (2*r/c)/np.sqrt(r**2-(c*p)**2) 

c = 8 # km/s
R = 6730 # radius in km

t= [] # travel times
t_udias = []
angles = np.linspace(0.01,0.99*np.pi/2,100) # (almost) all angles, just avoiding some singularities

for angle in angles:
    p = R*np.sin(angle)/c
    t.append(quad(integrand, p*c, R, args=(c,p))[0])
    t_udias.append((2/c)*np.sqrt(R**2-(p*c)**2))

plt.plot(angles,t,linewidth=2,label='numerical')
plt.plot(angles,t_udias,'--r',label='analytical')
plt.xlabel('take-off angle (radians)')
plt.ylabel('travel time (s)')
plt.legend()
plt.show()
```

```python
# integrand for the epicentral distance:
def integrand(r, c, p):
    return (2*p*c/r)/np.sqrt(r**2-(c*p)**2) 

delta= [] # epicentral distances
delta_udias = []
angles = np.linspace(0.01,0.99*np.pi/2,100) # (almost) all angles, just avoiding some singularities

for angle in angles:
    p = R*np.sin(angle)/c
    delta.append(quad(integrand, p*c, R, args=(c,p))[0])
    delta_udias.append(2*np.arccos(p*c/R)) # analytic solution for c = constant

plt.plot(angles,delta,linewidth=2,label='numerical')
plt.plot(angles,delta_udias,'--r',label='analytical')
plt.legend()
plt.xlabel('take-off angle (radians)')
plt.ylabel('epicentral distance (radians)')
plt.plot()
plt.show()
```

```python
plt.plot(delta,t,linewidth=2,label='numerical')
plt.ylabel('travel time (s)')
plt.xlabel('epicentral distance (radians)')
plt.plot()
plt.show()
```

This is not quite linear, because the speed is constant, but the epicentral distance is an arc. But how about plotting travel time $t$ as a function of ray length $s$? Could check Udias (analytic) or Aki/Richards (integral), or just use the sine rule: $$\frac{s}{\sin(\Delta)} = \frac{R}{\sin\left(\frac{\pi-\Delta}{2}\right)}.$$ Draw that situation, if this is not obvious!

```python
s = R*np.sin(delta)/np.sin((np.pi-np.array(delta))/2)
plt.plot(s,t)
plt.xlabel('Ray length (km)')
plt.ylabel('Travel time (s)')
plt.show()
```

... and we see that for a homogeneous sphere, the travel time is linear with the length of the ray. All is as expected! Udias derives another analytic solution when $v = v_0\left(\frac{R}{r}\right)^\alpha$. This has a singularity at the centre of the Earth ($r=0$), but has proved useful piece-wise in regions of the Earth (with specific values of $\alpha$):

```python
import numpy as np
import matplotlib.pyplot as plt

def t_vofr(R,alpha,v0,deltas):
    eta0 = R/v0
    return (2*eta0/(1+alpha))*np.sin((1+alpha)*deltas/2)

# making up some parameters for the crust:
alpha = 5
R = 100 # km
v0 = 5 #km/s
deltas = np.linspace(0,np.pi/10,100)
ts = t_vofr(R,alpha,v0,deltas)
plt.plot(deltas,ts)
plt.show()
```

## PREM

More realistic P- and S-wave speeds as a function of the depth are captured in the [Preliminary Reference Earth Model (PREM)](https://en.wikipedia.org/wiki/Preliminary_reference_Earth_model):

```python
import pandas as pd

url="http://ds.iris.edu/files/products/emc/data/PREM/PREM_ANISOTROPIC.csv"
c=pd.read_csv(url, usecols=[0,2], names=['radius', 'vP'], header=None)
vp =np.array(c['vP'])[::-1] # let's reverse the array so that 0 is the surface
r =np.array(c['radius'])/1000
plt.plot(r,vp)
plt.xlabel('Depth (km)')
plt.ylabel('Speed (km/s)')
plt.show()
```

In the Earth, the wavespeed is not continuous with depth. The strongest example of this is at the boundary between the outer and inner core. As a result, our approach of differentiation and integration to find travel times fail! Furthermore, such strong variations in wave speed lead to multi-valued travel time curves. More sophisticated schemes to estimate ray paths and travel times for seismic arrivals are explored in your ray tracing lab at the University of Auckland. In it, we use code based on the delay-time $\tau$, ray parameter ($p$) domain, originally proposed by Buland and Chapman (1983).
