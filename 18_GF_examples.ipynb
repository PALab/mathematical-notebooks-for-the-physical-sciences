```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# The Green's function for the diffusion equation in N dimensions:
def G(kappa,dim,t,x):
    return np.exp(-x**2/(4*kappa*t))/(4*np.pi*kappa*t)**(dim/2)
```

```python
from time import sleep
# the diffusivity of this infinite homogeneous space:
kappa = 10
    
# dimensions:
dims = [1,2,3]

# slices in space
x = np.linspace(-100,100,1000)

times = np.array([0.1,1,5,10])



#plt.savefig(str(dim)+'-D Gaussian2Delta.pdf')
fig = plt.figure()
for dim in dims:
    ax = fig.add_subplot(3,1, dim)
    plt.xlabel('Distance')
    plt.ylabel('Temperature')
    plt.title('Green function in '+str(dim)+'-D')
    for t in times:
        plt.plot(x,G(kappa,dim,t,x),label='time = '+str(t))

plt.legend()
plt.show()
    
    
```

```python
t = np.linspace(0,1000,1000)
x= 2
plt.figure()
for dim in dims:
    plt.plot(t,G(kappa,dim,t,x),label=str(dim)+'-D')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature at x='+str(x))
plt.legend()
#plt.savefig('DiffusioninTime.pdf')
plt.show()
```
